```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
import folium
from math import sin, cos, sqrt, atan2, radians

# Calculate distance function
def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance

# Coordinates for the points
lat1, lon1 = 28.56367, -80.57163
lat2, lon2 = 34.63377, -120.61066

# Calculate the distance
distance_coastline = calculate_distance(lat1, lon1, lat2, lon2)

# Calculate the midpoint for the marker
mid_lat = (lat1 + lat2) / 2
mid_lon = (lon1 + lon2) / 2

# Create a map centered around the midpoint
m = folium.Map(location=[mid_lat, mid_lon], zoom_start=5)

# Add markers for the points
folium.Marker(location=[lat1, lon1], popup='Point 1: 28.56367, -80.57163').add_to(m)
folium.Marker(location=[lat2, lon2], popup='Point 2: 34.63377, -120.61066').add_to(m)

# Add a line connecting the points
folium.PolyLine(locations=[[lat1, lon1], [lat2, lon2]], color='blue').add_to(m)

# Add a marker for the midpoint showing the distance
folium.Marker(
    location=[mid_lat, mid_lon],
    popup=f'Distance: {distance_coastline:.2f} km',
    icon=folium.Icon(color='red')
).add_to(m)

# Save the map to an HTML file
m.save('distance_map.html')

# Display the map
m
```
